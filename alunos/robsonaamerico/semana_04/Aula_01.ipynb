**Exportando para JSON — formato de APIs e sistemas modernos**

```python
import pandas as pd
import json

# ── Exportar o DataFrame para JSON ───────────────────────────
# orient="records"   → cada linha vira um objeto JSON independente
# force_ascii=False  → mantém ã é ç sem escapar (\u00e3 etc.)
# indent=2           → arquivo legível por humanos
df.to_json(
    "base_rh.json",
    orient="records",
    force_ascii=False,
    indent=2
)

print("✓ base_rh.json gerado")
```

```python
# ── Lendo de volta para confirmar ────────────────────────────
df_json = pd.read_json("base_rh.json")
print(f"Linhas lidas do JSON: {len(df_json)}")
```

```python
# ── Visualizar os 2 primeiros registros no formato JSON ───────
with open("base_rh.json", "r", encoding="utf-8") as f:
    amostra = json.load(f)
print("\nPrimeiros 2 registros no JSON:")
print(json.dumps(amostra[:2], ensure_ascii=False, indent=2))
```

**Por que isso importa?**

orient="records" gera o formato que APIs REST esperam — uma lista de objetos, não um objeto com arrays. force_ascii=False é essencial: sem ele, "Produção" vira "Produ\u00e7\u00e3o" no arquivo, o que quebra sistemas que não interpretam escape unicode. Em indústria, o JSON gerado aqui pode alimentar um dashboard web ou uma API interna do setor de RH.

```python
# ── Exportação avançada: uma aba por departamento ────────────
# ExcelWriter permite escrever múltiplas abas no mesmo arquivo
with pd.ExcelWriter("base_rh_por_depto.xlsx", engine="openpyxl") as writer:

    # Aba 1: dataset completo
    df.to_excel(writer, sheet_name="Todos", index=False)

    # Uma aba para cada departamento
    for depto in df["Departamento"].unique():
        df_depto = df[df["Departamento"] == depto]
        # Nome da aba limitado a 31 chars (limite do Excel)
        nome_aba = depto[:31]
        df_depto.to_excel(writer, sheet_name=nome_aba, index=False)
        print(f"  Aba '{nome_aba}': {len(df_depto)} registros")

print("\n✓ base_rh_por_depto.xlsx gerado")
```

```python
# ── Lendo de volta uma aba específica ────────────────────────
df_rh = pd.read_excel("base_rh_por_depto.xlsx", sheet_name="RH")
print(f"\nAba RH: {len(df_rh)} funcionários")
print(df_rh[["Nome", "Cargo", "Salario"]].head(4).to_string(index=False))
```

**Por que isso importa?**

O ExcelWriter como gerenciador de contexto (with) garante que o arquivo seja fechado corretamente — sem isso o Excel fica corrompido. O limite de 31 caracteres no nome da aba é uma restrição do Excel, não do pandas: se ultrapassar, o arquivo abre com erro. **Módulo datetime — trabalhando com datas**

"Turma, a coluna Data_Admissao chegou como string. O pandas não sabe que '13/08/2024' é uma data — pra ele é só texto. Se a gente não converter, não consegue calcular há quantos anos alguém está na empresa, não consegue filtrar quem foi admitido nos últimos 2 anos, não consegue agrupar por trimestre. Data é dado, não texto. Vamos resolver isso agora."

Convertendo datas e calculando tempo de casa

```python
import pandas as pd
print(pd.__version__)
from datetime import date, datetime, timedelta
```

```python
# ── 1. Entendendo strptime e strftime (módulo datetime puro) ──
data_texto = "20/07/1982"
```

```python
# strptime: converte string → objeto datetime
# %d = dia com zero à esquerda | %m = mês | %Y = ano 4 dígitos
data_obj = datetime.strptime(data_texto, "%d/%m/%Y")
print("Antes da conversão\n")
print(f"String original : {data_texto}")
print(f"Tipo original   : {type(data_texto)}\n")
print("Após conversão\n")
print(f"Objeto datetime : {data_obj}")
print(f"Tipo            : {type(data_obj)}")
```

```python
# strftime: converte objeto datetime → string formatada
formatado = data_obj.strftime("%B de %Y")   # %B = nome do mês
print(f"Formatado: {formatado}")
```

```python
# ── 2. timedelta: diferença entre datas ──────────────────────
hoje = datetime.today()
diferenca = hoje - data_obj          # resultado é um timedelta
print(f"\nHoje       : {hoje.strftime('%d/%m/%Y')}")
print(f"Diferença  : {diferenca.days} dias")
print(f"Em anos    : {diferenca.days / 365.25:.1f} anos")
```

```python
# ── 3. Aplicando no DataFrame inteiro ────────────────────────
# pd.to_datetime converte a coluna inteira — muito mais eficiente
# do que aplicar strptime linha a linha com .apply()
df["Data_Admissao"] = pd.to_datetime(
    df["Data_Admissao"],
    format="%d/%m/%Y",
    errors="coerce"   # datas inválidas → NaT (não quebra o pipeline)
)

print(f"\nTipo após conversão: {df['Data_Admissao'].dtype}")
print(f"Valores NaT gerados: {df['Data_Admissao'].isna().sum()}")
```

```python
# ── 4. Coluna de tempo de casa em anos ───────────────────────
hoje_ts = pd.Timestamp.today()
df["Tempo_Casa_Anos"] = (
    (hoje_ts - df["Data_Admissao"]).dt.days / 365.25
).round(1)

# ── 5. Resultado ─────────────────────────────────────────────
print("\nAmostra — Data convertida e tempo de casa:")
cols = ["Nome", "Data_Admissao", "Tempo_Casa_Anos"]
print(df[cols].head(6).to_string(index=False))
```

**Colunas derivadas de data + filtros por período**

Com a data convertida, extraímos atributos e aplicamos filtros — as operações mais pedidas em relatórios de RH industrial: headcount por ano de admissão, veteranos na empresa, admissões nos últimos 12 meses.

```python
# ── 1. Extraindo componentes da data com .dt ─────────────────
# .dt é o acessor de datetime no pandas — dá acesso a ano, mês, etc.
df["Ano_Admissao"]  = df["Data_Admissao"].dt.year
df["Mes_Admissao"]  = df["Data_Admissao"].dt.month
df["Dia_Semana"]    = df["Data_Admissao"].dt.day_name()   # nome do dia
df["Trimestre"]     = df["Data_Admissao"].dt.quarter

print("Colunas derivadas criadas:")
cols = ["Nome", "Data_Admissao", "Ano_Admissao", "Mes_Admissao", "Trimestre"]
print(df[cols].head(4).to_string(index=False))
```

```python
# ── 2. Faixa de tempo de casa (coluna categórica) ─────────────
def classificar_tempo(anos):
    if anos < 1:
        return "Menos de 1 ano"
    elif anos < 3:
        return "1 a 3 anos"
    elif anos < 5:
        return "3 a 5 anos"
    else:
        return "Mais de 5 anos"

df["Faixa_Tempo_Casa"] = df["Tempo_Casa_Anos"].apply(classificar_tempo)

print("\nDistribuição por faixa de tempo de casa:")
print(df["Faixa_Tempo_Casa"].value_counts().to_string())
```

```python
# ── 3. Filtros por período ────────────────────────────────────
hoje  = pd.Timestamp.today()
um_ano_atras = hoje - pd.DateOffset(years=1)
cinco_anos_atras = hoje - pd.DateOffset(years=5)

# Admitidos nos últimos 12 meses
novos = df[df["Data_Admissao"] >= um_ano_atras]

# Veteranos — mais de 5 anos de casa
veteranos = df[df["Data_Admissao"] < cinco_anos_atras]

print(f"\nAdmitidos nos últimos 12 meses : {len(novos)} funcionários")
print(f"Veteranos (5+ anos de casa)    : {len(veteranos)} funcionários")

# ── 4. Admissões por ano — panorama histórico ─────────────────
print("\nAdmissões por ano:")
print(
    df.groupby("Ano_Admissao")
    .size()
    .rename("Admissoes")
    .sort_index()
    .to_string()
)

# ── 5. Salário médio dos veteranos vs novos ───────────────────
print("\nSalário médio:")
print(f"  Novos (últimos 12 meses): R$ {novos['Salario'].mean():,.2f}")
print(f"  Veteranos (5+ anos)     : R$ {veteranos['Salario'].mean():,.2f}")
```

**Por que isso importa?**

pd.DateOffset(years=1) é mais correto que pd.Timedelta(days=365) para filtros de "último ano" — ele considera anos bissextos e meses com 28/30/31 dias. O insight do salário médio (veteranos ganham mais) é exatamente o tipo de análise que o gestor de RH vai pedir: "quem ficou ganhou mais com o tempo?" — e você responde com três linhas de código.
