**Estruturas**
**if (se)**

```python
# if é igual a palavra SE.
# Se a condição for atendida (verdadeira) o código dentro do if será executado.
# No exemplo abaixo, se a idade for maior ou igual a 18, a mensagem "Você é maior de idade." será exibida. Caso contrário, a mensagem "Você é menor de idade." será exibida.

idade = 18

if idade >= 18:
    print("Você é maior de idade.")
```

**elif (senão)**

```python
# No caso abaixo se a condição for atendida irá imprimir a mensagem "Você é maior de idade."
# Agora com o elif (senão) se a condição do if (se) não for atendida, o código abaixo do if que é o else será executado.
# 

idade = 17

if idade >= 18:
    print("Você é maior de idade.")
elif idade < 18:
    print("Você é menor de idade.")
```

**else (se/senão)**

```python
# Se umas das condições for atendida, o código dentro do if ou elif será executado. Caso contrário, o código dentro do else será executado.
# Sempre uma das condições deverá ser atendida.

nota = 5

if nota >= 8:
    print("Sua nota é excelente")
if nota >= 6 and nota < 8:
    print("Bom, mas pode melhorar")
elif nota >= 4 and nota < 6:
    print("Você ficou de recuperação")
else:
    print("Reprovado")
```
