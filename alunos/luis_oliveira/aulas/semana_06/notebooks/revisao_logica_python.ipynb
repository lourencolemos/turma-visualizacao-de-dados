## 🔷 Bloco 3 — Operadores

### Operadores Aritméticos (contas)

| Operador | Significado | Exemplo |
|----------|-------------|---------|
| `+` | Soma | `10 + 5 = 15` |
| `-` | Subtração | `10 - 3 = 7` |
| `*` | Multiplicação | `4 * 3 = 12` |
| `/` | Divisão | `10 / 4 = 2.5` |
| `//` | Divisão inteira | `10 // 4 = 2` |
| `%` | Resto da divisão | `10 % 3 = 1` |
| `**` | Potência | `2 ** 8 = 256` |

````python
# =============================================
# 🏭 CENÁRIO: Calculando indicadores de produção
# =============================================

pecas_produzidas = 1500
pecas_com_defeito = 37
horas_trabalhadas = 8
numero_operadores = 6
custo_por_peca = 12.50  # reais

# Peças boas
pecas_boas = pecas_produzidas - pecas_com_defeito
print(f"Peças boas: {pecas_boas}")

# Taxa de defeito (em %)
taxa_defeito = (pecas_com_defeito / pecas_produzidas) * 100
print(f"Taxa de defeito: {taxa_defeito:.2f}%")

# Produtividade por operador
produtividade = pecas_produzidas / numero_operadores
print(f"Peças por operador: {produtividade:.1f}")

# Produção por hora
pecas_por_hora = pecas_produzidas / horas_trabalhadas
print(f"Peças por hora: {pecas_por_hora:.1f}")

# Faturamento do turno
faturamento = pecas_boas * custo_por_peca
print(f"Faturamento do turno: R$ {faturamento:,.2f}")
````

````python
# Operadores de comparação — retornam True ou False
print("=== COMPARAÇÕES ===")
print(f"1500 > 1000? {1500 > 1000}")    # Maior que
print(f"37 < 50? {37 < 50}")            # Menor que
print(f"taxa_defeito <= 5? {taxa_defeito <= 5.0}")  # Menor ou igual
print(f"turno == 'Tarde'? {'Manhã' == 'Tarde'}")   # Igual
print(f"turno != 'Tarde'? {'Manhã' != 'Tarde'}")   # Diferente
````

````python
# Usando operadores lógicos: and, or, not

temperatura = 195.0
pressao = 3.2
valvula_aberta = True

# and = as DUAS condições precisam ser verdadeiras
if temperatura > 180 and pressao > 3.0:
    print("⚙️  Processo dentro do padrão")

# or = PELO MENOS UMA condição precisa ser verdadeira
if temperatura > 220 or pressao > 5.0:
    print("🚨 Risco de superaquecimento!")
else:
    print("✅ Temperatura e pressão normais")

# not = inverte o valor lógico
if not valvula_aberta:
    print("❌ Abra a válvula antes de iniciar!")
else:
    print("✅ Válvula aberta — pode iniciar")
````

## 🔷 Bloco 6 — Estruturas de Repetição

### 6.1 — Loop `for`
Usamos `for` quando sabemos **quantas vezes** queremos repetir algo.

````python
# =============================================
# 🏭 CENÁRIO: Relatório diário de produção
# =============================================

# Lista de produção por dia da semana
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]

print("=== PRODUÇÃO SEMANAL ===")

total = 0
for i in range(len(dias)):
    print(f"{dias[i]}: {producao_semanal[i]} peças")
    total += producao_semanal[i]  # acumula o total

media = total / len(dias)
print(f"\nTotal da semana: {total} peças")
print(f"Média diária: {media:.1f} peças")
````

````python
# Usando range() — gera sequências de números
print("Contagem de 1 a 5:")
for numero in range(1, 6):
    print(numero)

print("\nTabuada do 3:")
for i in range(1, 11):
    resultado = 3 * i
    print(f"3 x {i:2} = {resultado:3}")
````

````python
# for com if dentro — combinando tudo!
producao_semanal = [1450, 1380, 1520, 1490, 1600]
dias = ["Segunda", "Terça", "Quarta", "Quinta", "Sexta"]
meta_diaria = 1500

print("=== ANÁLISE DA META ===")
dias_acima = 0
dias_abaixo = 0

for i in range(len(dias)):
    producao = producao_semanal[i]
    if producao >= meta_diaria:
        print(f"✅ {dias[i]}: {producao} — META BATIDA")
        dias_acima += 1
    else:
        diferenca = meta_diaria - producao
        print(f"❌ {dias[i]}: {producao} — faltaram {diferenca} peças")
        dias_abaixo += 1

print(f"\nDias acima da meta: {dias_acima}")
print(f"Dias abaixo da meta: {dias_abaixo}")
````

### 6.2 — Loop `while`
Usamos `while` quando queremos repetir **enquanto uma condição for verdadeira**.

> ⚠️ **Cuidado com loops infinitos!** A condição precisa se tornar falsa em algum momento.

````python
# =============================================
# 🏭 CENÁRIO: Simulando uma esteira de produção
# =============================================

estoque_materias_primas = 100  # kg de matéria-prima
consumo_por_lote = 15          # kg consumidos por lote
lotes_produzidos = 0

print("=== SIMULAÇÃO DE PRODUÇÃO ===")
print(f"Estoque inicial: {estoque_materias_primas} kg\n")

while estoque_materias_primas >= consumo_por_lote:
    estoque_materias_primas -= consumo_por_lote
    lotes_produzidos += 1
    print(f"Lote {lotes_produzidos} produzido | Estoque restante: {estoque_materias_primas} kg")

print(f"\n⚠️  Matéria-prima insuficiente para novo lote")
print(f"Total de lotes produzidos: {lotes_produzidos}")
print(f"Sobra de material: {estoque_materias_primas} kg")
````

## 🔷 Bloco 8 — Funções

Funções são **blocos de código reutilizáveis**. Em vez de repetir o mesmo código, você cria uma função e chama ela quando precisar.

```python
def nome_da_funcao(parametro1, parametro2):
    # código aqui
    return resultado
```

````python
# =============================================
# 🏭 CENÁRIO: Funções para indicadores de qualidade
# =============================================

def calcular_taxa_defeito(total_produzido, total_defeituoso):
    """Calcula a taxa de defeito em porcentagem."""
    if total_produzido == 0:
        return 0
    taxa = (total_defeituoso / total_produzido) * 100
    return taxa


def classificar_qualidade(taxa_defeito):
    """Classifica a qualidade com base na taxa de defeito."""
    if taxa_defeito <= 1.0:
        return "EXCELENTE"
    elif taxa_defeito <= 3.0:
        return "BOM"
    elif taxa_defeito <= 5.0:
        return "ATENÇÃO"
    else:
        return "CRÍTICO"


def gerar_relatorio_turno(turno, operador, pecas, defeitos):
    """Gera um relatório completo do turno."""
    taxa = calcular_taxa_defeito(pecas, defeitos)
    classificacao = classificar_qualidade(taxa)
    pecas_boas = pecas - defeitos
    
    print(f"\n{'='*40}")
    print(f"  RELATÓRIO — TURNO {turno.upper()}")
    print(f"{'='*40}")
    print(f"  Operador      : {operador}")
    print(f"  Peças prod.   : {pecas}")
    print(f"  Peças boas    : {pecas_boas}")
    print(f"  Defeituosas   : {defeitos}")
    print(f"  Taxa defeito  : {taxa:.2f}%")
    print(f"  Classificação : {classificacao}")
    print(f"{'='*40}")


# Chamando as funções
gerar_relatorio_turno("Manhã", "Ana Silva", 520, 8)
gerar_relatorio_turno("Tarde", "Bruno Costa", 495, 28)
gerar_relatorio_turno("Noite", "Carla Souza", 485, 4)
````

### ✏️ Atividade Acompanhada 4 — Crie sua função
Crie uma função que calcule o OEE (Overall Equipment Effectiveness) — um dos indicadores mais importantes na indústria:

````python
# ✏️ ATIVIDADE ACOMPANHADA 4
# OEE = Disponibilidade × Performance × Qualidade × 100
# Cada fator vai de 0 a 1 (ex: 0.85 = 85%)

def calcular_oee(disponibilidade, performance, qualidade):
    """
    Calcula o OEE (Overall Equipment Effectiveness)
    
    Parâmetros:
        disponibilidade: tempo disponível / tempo planejado (0 a 1)
        performance: velocidade real / velocidade ideal (0 a 1)
        qualidade: peças boas / total produzido (0 a 1)
    """
    # TODO: calcule o OEE
    oee = disponibilidade * performance * qualidade * 100
    return oee


# Testando a função
oee_resultado = calcular_oee(
    disponibilidade=0.90,  # máquina disponível 90% do tempo
    performance=0.95,      # operando a 95% da velocidade ideal
    qualidade=0.98         # 98% de peças boas
)

print(f"OEE calculado: {oee_resultado:.1f}%")

# OEE de classe mundial é ≥ 85%
if oee_resultado >= 85:
    print("🏆 Equipamento de CLASSE MUNDIAL!")
else:
    print("📈 Há oportunidade de melhoria")
````

### 🔹 Atividade Individual 1 — Calculadora de Salário
**Dificuldade:** ⭐☆☆ (Fácil)

````python
# ============================================================
# 📝 ATIVIDADE INDIVIDUAL 1 — Calculadora de Salário
# ============================================================
# 
# ENUNCIADO:
# Um operador recebe R$ 18,50 por hora normal.
# Horas extras (acima de 44h semanais) valem 50% a mais.
#
# Crie um programa que:
# 1. Peça o nome do funcionário
# 2. Peça as horas trabalhadas na semana
# 3. Calcule o salário semanal (com horas extras se houver)
# 4. Exiba um comprovante formatado
#
# DICA:
# - Salário normal = horas normais x R$ 18,50
# - Horas extras = horas trabalhadas - 44 (se > 44)
# - Valor hora extra = 18,50 x 1,5
# ============================================================

# SEU CÓDIGO AQUI:

# =============================================
# CENÁRIO: Sistema de cálculo salarial
# =============================================

print("=== CALCULADORA DE SALÁRIO SEMANAL ===")
print()

# Entrada de dados
funcionario = input("Nome do funcionário: ")

# Tratamento de erro
try:
    
    horas_trabalhadas = float(input("Horas trabalhadas na semana: "))
    
    # Valores fixos
    valor_hora = 18.50
    valor_hora_extra = valor_hora * 1.5

    # Verificação de horas extras
    if horas_trabalhadas > 44:
        
        horas_normais = 44
        horas_extras = horas_trabalhadas - 44
        
    else:
        
        horas_normais = horas_trabalhadas
        horas_extras = 0

    # Cálculos
    salario_normal = horas_normais * valor_hora
    salario_extra = horas_extras * valor_hora_extra
    salario_total = salario_normal + salario_extra

    # Comprovante
    print()
    print("========== COMPROVANTE ==========")
    print(f"Funcionário: {funcionario}")
    print(f"Horas normais: {horas_normais}")
    print(f"Horas extras: {horas_extras}")
    print(f"Salário normal: R$ {salario_normal:.2f}")
    print(f"Salário extra: R$ {salario_extra:.2f}")
    print(f"Salário total: R$ {salario_total:.2f}")
    print("=================================")

except:
    
    print("Erro: digite um número válido para as horas.")
````

### 🔶 Atividade em Grupo 1 — Sistema de Registro de Produção
**Grupo:** 3-4 pessoas | **Dificuldade:** ⭐⭐☆

````python
# ============================================================
# 👥 ATIVIDADE EM GRUPO 1 — Sistema de Registro de Produção
# ============================================================
#
# ENUNCIADO:
# Vocês devem criar um mini-sistema de registro de produção.
# O sistema deve:
#
# 1. Usar um while para manter o programa rodando
# 2. Apresentar um menu com as opções:
#    [1] Registrar produção de um turno
#    [2] Ver todos os registros
#    [3] Ver estatísticas (total, maior, menor, média)
#    [4] Sair
#
# 3. Armazenar os registros em uma lista de dicionários
# 4. Cada registro deve conter:
#    - nome do operador
#    - turno (Manhã/Tarde/Noite)
#    - quantidade produzida
#    - quantidade com defeito
#    - taxa de defeito (calculada automaticamente)
#
# DICA: Dividam as tarefas! Uma pessoa faz o menu,
#       outra faz o registro, outra faz as estatísticas.
# ============================================================

# CÓDIGO DO GRUPO AQUI:


    
````

````python
# ============================================================
# 👥 ATIVIDADE EM GRUPO 1 — Sistema de Registro de Produção
# ============================================================

# Lista que armazenará os registros
registros = []

# ============================================================
# FUNÇÃO — Registrar produção
# ============================================================

def registrar_producao():

    print("\n===== REGISTRO DE PRODUÇÃO =====")

    operador = input("Nome do operador: ").strip()

    turno = input(
        "Turno (Manhã/Tarde/Noite): "
    ).strip()

    quantidade = int(
        input("Quantidade produzida: ").strip()
    )

    defeitos = int(
        input("Quantidade com defeito: ").strip()
    )

    # Cálculo da taxa de defeito
    taxa_defeito = (
        defeitos / quantidade
    ) * 100

    # Criação do registro
    registro = {
        "operador": operador,
        "turno": turno,
        "quantidade": quantidade,
        "defeitos": defeitos,
        "taxa_defeito": taxa_defeito
    }

    # Salva na lista
    registros.append(registro)

    print("\nstro salvo com sucesso!")

# ============================================================
# 📌 FUNÇÃO — Mostrar registros
# ============================================================

def mostrar_registros():

    print("\n===== TODOS OS REGISTROS =====")

    # Verifica se existem registros
    if len(registros) == 0:

        print("Nenhum registro encontrado.")
        return

    # Mostra os registros salvos
    for registro in registros:

        print("-----------------------------------")

        print(
            f"Operador: "
            f"{registro['operador']}"
        )

        print(
            f"Turno: "
            f"{registro['turno']}"
        )

        print(
            f"Produção: "
            f"{registro['quantidade']}"
        )

        print(
            f"Defeitos: "
            f"{registro['defeitos']}"
        )

        print(
            f"Taxa de defeito: "
            f"{registro['taxa_defeito']:.2f}%"
        )

# ============================================================
# FUNÇÃO — Estatísticas
# ============================================================

def ver_estatisticas():

    print("\n===== ESTATÍSTICAS =====")

    # Verifica se existem registros
    if len(registros) == 0:

        print("Nenhum registro encontrado.")
        return

    # Lista apenas das produções
    producoes = []

    # Percorre os registros
    for registro in registros:

        producoes.append(
            registro["quantidade"]
        )

    # Cálculos
    total = sum(producoes)

    maior = max(producoes)

    menor = min(producoes)

    media = total / len(producoes)

    # Exibição
    print(f"Total produzido: {total}")

    print(f"Maior produção: {maior}")

    print(f"Menor produção: {menor}")

    print(f"Média de produção: {media:.2f}")

# ============================================================
# MENU PRINCIPAL
# ============================================================

while True:

    print("\n================================")
    print("SISTEMA DE PRODUÇÃO")
    print("================================")

    print("[1] Registrar produção")
    print("[2] Ver registros")
    print("[3] Ver estatísticas")
    print("[4] Sair")

    opcao = input(
        "\nEscolha uma opção: "
    ).strip()

    # ========================================================
    # OPÇÃO 1 — Registrar produção
    # ========================================================

    if opcao == "1":

        registrar_producao()

    # ========================================================
    # OPÇÃO 2 — Mostrar registros
    # ========================================================

    elif opcao == "2":

        mostrar_registros()

    # ========================================================
    # OPÇÃO 3 — Estatísticas
    # ========================================================

    elif opcao == "3":

        ver_estatisticas()

    # ========================================================
    # OPÇÃO 4 — Encerrar sistema
    # ========================================================

    elif opcao == "4":

        print("\nSistema encerrado.")
        break

    # ========================================================
    # OPÇÃO INVÁLIDA
    # ========================================================

    else:

        print("\nOpção inválida!")
````

````python
# ============================================================
# 👥 ATIVIDADE EM GRUPO 2 — Dashboard de Qualidade
# ============================================================

dados_semana = [
    {"dia": "Segunda",  "turno": "Manhã",  "operador": "Ana",    "pecas": 520, "defeitos": 8},
    {"dia": "Segunda",  "turno": "Tarde",  "operador": "Bruno",  "pecas": 495, "defeitos": 18},
    {"dia": "Segunda",  "turno": "Noite",  "operador": "Carla",  "pecas": 485, "defeitos": 5},
    {"dia": "Terça",    "turno": "Manhã",  "operador": "Daniel", "pecas": 510, "defeitos": 12},
    {"dia": "Terça",    "turno": "Tarde",  "operador": "Ana",    "pecas": 530, "defeitos": 6},
    {"dia": "Terça",    "turno": "Noite",  "operador": "Bruno",  "pecas": 340, "defeitos": 25},
    {"dia": "Quarta",   "turno": "Manhã",  "operador": "Carla",  "pecas": 560, "defeitos": 4},
    {"dia": "Quarta",   "turno": "Tarde",  "operador": "Daniel", "pecas": 475, "defeitos": 30},
    {"dia": "Quarta",   "turno": "Noite",  "operador": "Ana",    "pecas": 500, "defeitos": 7},
    {"dia": "Quinta",   "turno": "Manhã",  "operador": "Bruno",  "pecas": 545, "defeitos": 9},
    {"dia": "Quinta",   "turno": "Tarde",  "operador": "Carla",  "pecas": 490, "defeitos": 11},
    {"dia": "Quinta",   "turno": "Noite",  "operador": "Daniel", "pecas": 465, "defeitos": 6},
    {"dia": "Sexta",    "turno": "Manhã",  "operador": "Ana",    "pecas": 580, "defeitos": 3},
    {"dia": "Sexta",    "turno": "Tarde",  "operador": "Bruno",  "pecas": 500, "defeitos": 14},
    {"dia": "Sexta",    "turno": "Noite",  "operador": "Carla",  "pecas": 455, "defeitos": 8},
]

# ============================================================
# FUNÇÃO 1 — Resumo por turno
# ============================================================

def resumo_turnos(dados):

    print("\n========== RESUMO POR TURNO ==========")

    turnos = {}

    # Organiza os dados por turno
    for registro in dados:

        turno = registro["turno"]

        if turno not in turnos:

            turnos[turno] = {
                "pecas": 0,
                "defeitos": 0,
                "registros": 0
            }

        turnos[turno]["pecas"] += registro["pecas"]
        turnos[turno]["defeitos"] += registro["defeitos"]
        turnos[turno]["registros"] += 1

    # Exibição
    for turno, info in turnos.items():

        media = info["pecas"] / info["registros"]

        print(f"\nTurno: {turno}")
        print(f"Média de produção: {media:.2f} peças")
        print(f"Total de defeitos: {info['defeitos']}")

# ============================================================
# FUNÇÃO 2 — Melhor e pior dia
# ============================================================

def melhor_pior_dia(dados):

    print("\n========== MELHOR E PIOR DIA ==========")

    dias = {}

    for registro in dados:

        dia = registro["dia"]

        if dia not in dias:

            dias[dia] = {
                "pecas": 0,
                "defeitos": 0
            }

        dias[dia]["pecas"] += registro["pecas"]
        dias[dia]["defeitos"] += registro["defeitos"]

    taxas = {}

    for dia, info in dias.items():

        taxa = (info["defeitos"] / info["pecas"]) * 100
        taxas[dia] = taxa

    melhor = min(taxas, key=taxas.get)
    pior = max(taxas, key=taxas.get)

    print(f"Melhor dia: {melhor} ({taxas[melhor]:.2f}% defeitos)")
    print(f"Pior dia: {pior} ({taxas[pior]:.2f}% defeitos)")

# ============================================================
# FUNÇÃO 3 — Ranking de operadores
# ============================================================

def ranking_operadores(dados):

    print("\n========== RANKING DE OPERADORES ==========")

    operadores = {}

    for registro in dados:

        operador = registro["operador"]

        if operador not in operadores:

            operadores[operador] = {
                "pecas": 0,
                "defeitos": 0
            }

        operadores[operador]["pecas"] += registro["pecas"]
        operadores[operador]["defeitos"] += registro["defeitos"]

    ranking = []

    for operador, info in operadores.items():

        eficiencia = (
            (info["pecas"] - info["defeitos"]) / info["pecas"]
        ) * 100

        ranking.append((operador, eficiencia))

    ranking.sort(key=lambda x: x[1], reverse=True)

    posicao = 1

    for operador, eficiencia in ranking:

        print(f"{posicao}º - {operador}: {eficiencia:.2f}% eficiência")
        posicao += 1

# ============================================================
# FUNÇÃO 4 — Alertas de qualidade
# ============================================================

def alertas(dados):

    print("\n========== ALERTAS ==========")

    encontrou_alerta = False

    for registro in dados:

        taxa = (registro["defeitos"] / registro["pecas"]) * 100

        if taxa > 3:

            encontrou_alerta = True

            print(
                f"⚠️ {registro['dia']} | "
                f"{registro['turno']} | "
                f"{registro['operador']} → "
                f"{taxa:.2f}% defeitos"
            )

    if not encontrou_alerta:

        print("Nenhum alerta encontrado.")

# ============================================================
# FUNÇÃO 5 — Mini gráfico com barra lateral
# ============================================================

def mini_grafico(dados):

    print("\n========== MINI GRÁFICO ==========")

    producao_dias = {}

    # Soma produção por dia
    for registro in dados:

        dia = registro["dia"]

        if dia not in producao_dias:

            producao_dias[dia] = 0

        producao_dias[dia] += registro["pecas"]

    # Gráfico de barras lateral
    for dia, total in producao_dias.items():

        barras = "█" * (total // 100)

        print(f"{dia:<10} | {barras} {total}")
# ============================================================
# EXECUÇÃO DO DASHBOARD
# ============================================================

print("======================================")
print("DASHBOARD DE QUALIDADE INDUSTRIAL")
print("======================================")

resumo_turnos(dados_semana)

melhor_pior_dia(dados_semana)

ranking_operadores(dados_semana)

alertas(dados_semana)

mini_grafico(dados_semana)

print("\nRelatório finalizado com sucesso!")
````
