```python
# 1. Estrutura de lista
# Instrução: Crie uma lista chamada cidades com 4 nomes de cidades que você conhece. Depois, exiba a lista inteira no terminal.

cidades = ['Florianopolis','Barcelona','Jaragua do Sul', 'Caracas']

print(cidades)
```

```python
## 2. Descobrir o índice de uma lista
# Instrução: Dada a lista faça o programa descobrir e exibir em qual posição está o animal "coelho".

animais = ["cachorro", "gato", "papagaio", "coelho", "hamster"]

i = animais.index("coelho") #Criamos a variavel index para achar a posicao do coelho na lista, chamando o coelho atraves dos ()

print('Nosso coelho esta na posicao '+ str(i)) # O print permite revelar sua posicao quando usamos a expresao + str (por que e um item de tipo string) e executando o index nos ()
```

```python
### 3. Adicionar ou remover
# Instrução: Crie uma lista vazia chamada compras. Adicione 3 itens nela usando append().
# Depois, remova o último item usando pop(). Exiba a lista após cada etapa.

compras = []
for i in range (3):
    novos_produtos = str(input('Porfavor adicione os novos produtos ao estoque'))
    compras.append(novos_produtos)

print('novos produtos', compras)

produto_vendido = compras.pop(2)
print('produto fora do estoque', produto_vendido)
print('produtos disponiveis', compras)
```

```python

#### 4. Usar TRY e EXCEPT
# Instrução: Crie uma lista com 3 números de sua escolha. 
# Peça para o usuário digitar um índice. Tente mostrar o elemento desse índice. 
# Se o índice não existir, capture o erro e exiba "Índice inválido!".

posicoes_astrales = ['777','969','1313']

pesquisa_posicoes = int(input('Digite o indice da lista (somente um numero de 0 ao 2):  '))

try:
    print(f'Voce digitou o indice {pesquisa_posicoes} ')
    print(f'A posicao astral escolhada por voce foi {posicoes_astrales[pesquisa_posicoes]}') #Reparei que sem o f-string nao seria possivel mostrar o valor das listas e os diccionarios dentro do print
    print(f'indice valido')

except IndexError:  #Da um nome ao possivel erro, existe outro tipos de erro que daria para avisar ao except colocar tipo valueError quando e letra por numero
    print(f'indice invalido')
```

```python
##### 5. Métodos para obter informações da lista
# Instrução: Dada a lista notas = [8, 7, 9, 6, 8, 7, 10, 8], use os métodos/funções corretos para exibir:
# Quantos alunos têm nota (tamanho da lista)
#1. A maior nota
#2. A menor nota
#3. Quantas vezes a nota 8 aparece

notas = [8, 7, 9, 6, 8, 7, 10, 8]
print(f'Na turma temos {len(notas)}') # {} serve para chamar variaves, funcoes, metodos
print(f'Na turma temos {max(notas)}')
print(f'Na turma temos {min(notas)}')

#Como sei quantas vezes aparece 8?
print(f'A nota 8 aparece {notas.count(8)}')


```

```python
###### 6. Juntar listas numa lista
#Instrução: Crie duas listas: lista_a = [1, 2, 3] e lista_b = [4, 5, 6]. 
# Una as duas em uma nova lista chamada lista_unida e exiba o resultado.

lista_a = [1, 2, 3]
lista_b = [4, 5, 6]
# Opcao A
lista_unida = lista_a + lista_b
print(lista_unida)
```

```python
# Opcao B
lista_a = [1, 2, 3]
lista_b = [4, 5, 6]
lista_a.extend(lista_b)
print(lista_a)
lista_unida = lista_a
print(lista_unida)
```

```python
# Opcao C
lista_a = [1, 2, 3]
lista_b = [4, 5, 6]
lista_unida = lista_a.append(lista_b)
print(lista_a)
lista_unida= lista_a
```

```python
####### 7. Ordenar lista
# Instrução: Dada a lista numeros = [15, 3, 8, 1, 20, 7], ordene em ordem crescente e exiba. 
# Depois, ordene em ordem decrescente e exiba.

numeros = [15, 3, 8, 1, 20, 7]
numeros.sort()
print(numeros)

numeros.sort(reverse=True)
print(numeros)
```

```python
#### #### 8. Usar método JOIN
#Instrução: Dada a lista palavras = ["hoje", "é", "um", "bom", "dia" , "para", "estudar" ]
# Junte todas as palavras em uma única frase, separadas por espaço, e exiba.

palavras = ["hoje", "é", "um", "bom", "dia","para", "estudar" ]

palavras = " ".join(palavras) # o que vai estar entre aspas vai ser espaco em branco para que esses itens possam estar numa leitura sem separador
print(palavras)
```
