# Passo 1: carregue o CSV

```python
#PRIMEIRO LUGAR PRECISAMOS BAIXAR As BIBLIOTECA PANDAS e R PARA CONSEGUIR PUXAR NOSSO ARQUIVO NO LOCAL
from unicodedata import decimal

import pandas as pd
import re
print('pandas versao:', pd.__version__)
print('Bibliotecas importadas com sucesso!')

#Comando: Criamos uma variavel para que armazena nossa rota de acesso ao arquivo principal
PASTA= r'C:\Users\miran\OneDrive\Desktop\turma-visualizacao-de-dados\alunos\isaac_Trenard\semana_04_ANOTAÇÕES\base_rh.csv'
print('Base de dados carregada com sucesso!')


df= pd.read_csv(
    PASTA,
    sep=';',           
    encoding='cp1252',
    decimal=',' )
print('DF Decodificado com sucesso!')
```

# Passo 2: diagnóstico

```python
print(f'  Total de linhas  : {df.shape[0]}')
print(f'  Total de colunas : {df.shape[1]}')
print('\n Diagnostico de Tipos de dados de cada coluna')
df.info()
print('\n Encontramos que cada coluna tem os seguinte valores nulos') 
print('\n Dado faltante reprensentado= 1, e dado completo ou preenchido= 0')
print(f'\n{(df.isnull().sum())}')
print(f'\n Colunas do dataset:')
print(f'\n{df.columns.tolist()}')
print('\n Amostra dos dados:')
print('\n Primeiros dados na base de dados ')
df.head()

```

# Passo 3: converter datas e criar colunas derivadas

```python
from datetime import datetime, timedelta
data_str = '13/08/2024' 
data_objeto = datetime.strptime(data_str, '%d/%m/%Y')

print(f'Valor original da coluna DATA_ADMISSAO {data_str}')
print(f'Valor convertido da coluna DATA_ADMISSAO {data_objeto}')
print(f'Verificacao tipo de dado da coluna Data_Admissao {type(data_objeto)}')

print(f'Extrair elemento separado Ano : {data_objeto.year}')
print(f'Extrair elemento separado Mês : {data_objeto.month}')
print(f'Extrair elemento separado Dia : {data_objeto.day}')

print(f'Formato original da coluna Data Admissao {df['Data_Admissao']}.dtype')

df_ajustado = df.copy()
df_ajustado['Data_Admissao'] = pd.to_datetime(df_ajustado['Data_Admissao'], format='%d/%m/%Y', errors='coerce')
print(f'DATA DE ADMISSAO AJUSTADA PARA : {df_ajustado['Data_Admissao'].dtype}')
hoje = datetime.today ()
periodo_de_tempo = hoje - data_objeto

print(f'Data de admissão : {data_str}')
print(f'Hoje: {hoje.strftime("%d/%m/%Y")}')
print(f'Periodo de tempo: {periodo_de_tempo.days} dias')
print(f'Em anos: {periodo_de_tempo.days / 365.25:.1f} anos')

df_ajustado['Ano_Admissao'] = df_ajustado['Data_Admissao'].dt.year
df_ajustado['Mes_admissao'] = df_ajustado['Data_Admissao'].dt.month
df_ajustado['Trimestre'] = df_ajustado['Data_Admissao'].dt.quarter
df_ajustado['Tempo_na_empresa'] = ((hoje - df_ajustado['Data_Admissao']).dt.days/365.25 ).round(1)

df_ajustado.head()

```

# Passo 4: groupby — maior e menor média de tempo de casa

```python
media_cargo = df_ajustado.groupby('Cargo')['Tempo_na_empresa'].mean().round(1)
print('Tempo médio na empresa por cargo:')
print(media_cargo)

maior_media = media_cargo.idxmax()
menor_media = media_cargo.idxmin()
print(f'Tempo minimo por cargo na empresa {menor_media} - {media_cargo[maior_media]} anos')
print(f'Tempo maximo por cargo na empresa {maior_media}- {media_cargo[menor_media]} anos')
```
