```python
# ============================================================
# 👥 ATIVIDADE EM GRUPO 1 — Sistema de Registro de Produção
# ============================================================
#
# ENUNCIADO:
# Vocês devem criar um mini-sistema de registro de produção.
# O sistema deve:
#
# 1. Usar um while para manter o programa rodando
# 2. Apresentar um menu com as opções:
#    [1] Registrar produção de um turno
#    [2] Ver todos os registros
#    [3] Ver estatísticas (total, maior, menor, média)
#    [4] Sair
#
# 3. Armazenar os registros em uma lista de dicionários
# 4. Cada registro deve conter:
#    - nome do operador
#    - turno (Manhã/Tarde/Noite)
#    - quantidade produzida
#    - quantidade com defeito
#    - taxa de defeito (calculada automaticamente)
#
# DICA: Dividam as tarefas! Uma pessoa faz o menu,
#       outra faz o registro, outra faz as estatísticas.
# ============================================================

# CÓDIGO DO GRUPO AQUI:

registros = []  # lista para armazenar os registros

def registrar_producao():
    operador = input("Nome do operador: ")
    turno = input("Turno (Manhã/Tarde/Noite):")
    qtd_produzida = int(input("quantidade produzida:"))
    qtd_defeito = int(input("quantidade com defeito:"))
    taxa_defeito = qtd_defeito/qtd_produzida * 100
    registros.append({"operador":operador,"turno":turno,"qtd_produzida":qtd_produzida,"qtd_defeito":qtd_defeito, "taxa_defeito":taxa_defeito})
    pass

def ver_registros():
    for registro in registros:
        print(f"Operador: {registro['operador']}")
        print(f"Turno: {registro['turno']}")
        print(f"Quantidade produzida: {registro['qtd_produzida']}")
        print(f"Quantidade com defeito: {registro['qtd_defeito']}")
        print(f"Taxa de defeito: {registro['taxa_defeito']:.2f}%")
        print("-" * 30)
    pass

def ver_estatisticas():
    producao = []
    pecas_defeito = []
    maior = 0
    menor = registros[0]['qtd_produzida']
    media = 0

    for registro in registros:
        
        producao.append(registro['qtd_produzida'])
        pecas_defeito.append(registro['qtd_defeito'])
        if maior > registro['qtd_produzida']:
            maior = registro['qtd_produzida']
        if menor < registro['qtd_produzida']:
            menor = registro['qtd_produzida']
    total = sum(producao)
    total_defeito = sum(pecas_defeito)
    media = total / len(registros)
    print(f"O total de peças é: {total}")
    print(f"O total de peças com defeito é: {total_defeito}")
    print(f"O maior registro é: {maior}")
    print(f"O menor registro é: {menor}")
    print(f"A média de peças é:{media}")
    pass 

# Menu principal
print("Sistema iniciado! Implemente o menu aqui.")
while True:
    print("====== Menu =======")
    print("[1] Registrar produção de um turno")
    print("[2] Ver todos os registros")
    print("[3] Ver estatísticas (total, maior, menor, média)")
    print("[4] Sair")
    escolha = int(input("Escolha a opção:"))
    if escolha == 1:
        registrar_producao()
    elif escolha == 2:
        ver_registros()
    elif escolha == 3:
        ver_estatisticas()
    elif escolha == 4:
        break
    else:
        print("escolha inválida")
```
